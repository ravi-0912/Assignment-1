```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
path = r"C:\Users\verma\Desktop\Selenium\chromedriver.exe"
```

```python
driver = webdriver.Chrome(path)
```

```python
driver.get('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
content=driver.page_source
soup= BeautifulSoup(content)
names=soup.findAll('td',attrs={'class':'titleColumn'})
print(len(names))
year=soup.findAll('span',attrs={'class':'secondaryInfo'})
print(len(year))
rating=soup.findAll('td',attrs={'class':'ratingColumn imdbRating'})
print(len(rating))
```

```python
names
```

```python
year
```

```python
rating
```

```python
Names=[]
Year=[]
Rating=[]
```

```python
for i in names:
     for k in i.find_all('a'):
        Names.append(k)
Names
```

```python
Movies=[]
```

```python
for i in Names:
    Movies.append(i.text)
Movies
```

```python
for i in year:
    Year.append(i.text)
Year
```

```python
for i in rating:
    Rating.append(i.text)
Rating
```

```python
df2= pd.DataFrame(Rating)
df2[0]=df2[0].str.replace('\n',' ')

df2=df2.rename({0:'Rating'},axis=1)
df2
```

```python
IMDB_Movies= pd.DataFrame({})
IMDB_Movies['Movies Name']=Movies
IMDB_Movies['Year']=Year
IMDB_Movies['Rating']=df2
```

```python
IMDB_Movies
```

```python
IMDB_Movies=IMDB_Movies.iloc[0:99]
IMDB_Movies
```

# Saving IMDB Top 100 Indian Movies Data into csv file
