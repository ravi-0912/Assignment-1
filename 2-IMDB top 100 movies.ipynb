```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
path = r"C:\Users\verma\Desktop\Selenium\chromedriver.exe"
```

```python
driver = webdriver.Chrome(path)
```

```python
driver.get('https://www.imdb.com/list/ls091520106/')
```

```python
content=driver.page_source
soup= BeautifulSoup(content)
names=soup.findAll('h3',attrs={'class':'lister-item-header'})
print(len(names))
year=soup.findAll('span',attrs={'class':'lister-item-year text-muted unbold'})
print(len(year))
rating=soup.findAll('div',attrs={'class':'ipl-rating-star small'})
print(len(rating))
```

```python
Names=[]
Year=[]
Rating=[]
```

```python
for i in names:
     for k in i.find_all('a'):
        Names.append(k)
Names
```

```python
Movies=[]
```

```python
for i in Names:
    Movies.append(i.text)
Movies
```

```python
df=pd.DataFrame(Movies)
df=df.rename({0:'Movie Name'},axis=1)
df
```

```python
for i in year:
    Year.append(i.text)
Year
```

```python
df1=pd.DataFrame(Year)
df1[0]=df1[0].str.replace('I', ' ')
df1=df1.rename({0:'Year'},axis=1)
df1
```

```python
for i in rating:
    Rating.append(i.text)
Rating
```

```python
df2= pd.DataFrame(Rating)
df2[0]=df2[0].str.replace('\n',' ')
df2=df2.rename({0:'Rating'},axis=1)
df2
```

```python
IMDB_Movies= pd.DataFrame({})
IMDB_Movies['Movies']=Movies
IMDB_Movies['Year']=Year
IMDB_Movies['Rating']=df2
```

```python
IMDB_Movies
```

# Saving IMDB Top Rated Movies Data in csv file

```python
IMDB_Movies.to_csv("IMDB_Top_100_Movies.csv")
```
